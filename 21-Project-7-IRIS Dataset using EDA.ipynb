```python
# Check column names
print(IRIS.columns)

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Subplot 1: Petal Length
sns.violinplot(ax=axes[0, 0], x='species', y='petal_length', data=IRIS)
axes[0, 0].set_title("Petal Length")

# Subplot 2: Petal Width
sns.violinplot(ax=axes[0, 1], x='species', y='petal_width', data=IRIS)
axes[0, 1].set_title("Petal Width")

# Subplot 3: Sepal Length
sns.violinplot(ax=axes[1, 0], x='species', y='sepal_length', data=IRIS)
axes[1, 0].set_title("Sepal Length")

# Subplot 4: Sepal Width
sns.violinplot(ax=axes[1, 1], x='species', y='sepal_width', data=IRIS)
axes[1, 1].set_title("Sepal Width")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#Dasboard

sns.set_style('darkgrid')
f, axes = plt.subplots(2,2, figsize=(15,15))

k1 = sns.boxenplot(x="species", y="petal_length", data=IRIS, ax=axes[0,0])
k2 = sns.violinplot(x="species", y="petal_length", data=IRIS, ax=axes[0,1])
k3 = sns.stripplot(
    x="species", 
    y="sepal_length", 
    data=IRIS, 
    jitter=True, 
    edgecolor='gray', 
    size=8, 
    palette='winter',  # Fixed typo: 'paletter' -> 'palette'
    orient='v', 
    ax=axes[1, 0]
)
axes[1, 1].hist(IRIS["petal_length"], bins=100, color='blue', edgecolor='black')
plt.show()
```

```python
#Stacked Histogram
IRIS['species'] = IRIS['species'].astype('category')
list1 = list()
mylabels = list()
for gen in IRIS.species.cat.categories:
    list1.append(IRIS[IRIS.species==gen].sepal_length)
    mylabels.append(gen)
h = plt.hist(list1, bins=30, stacked=True, rwidth=1, label=mylabels)
plt.legend()
plt.show()
```

```python
# Area Plot
IRIS.plot.area(y=['sepal_length','sepal_width','petal_length','petal_width'], alpha = 0.4, figsize=(12,6))
plt.show()
```
